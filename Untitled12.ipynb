```python
from langchain.document_loaders import PyMuPDFLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.chains import RetrievalQA
from transformers import pipeline, AutoModelForSeq2SeqLM, AutoTokenizer
from langchain.llms import HuggingFacePipeline
import os
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
loader = PyMuPDFLoader("MLLec2Part1.pdf")  # Replace with your file name
documents = loader.load()

# Split the text into smaller chunks
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
docs = text_splitter.split_documents(documents)
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectorstore = FAISS.from_documents(docs, embeddings)
```

```python
model_name = "google/flan-t5-base"  # Lightweight and free
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

pipe = pipeline("text2text-generation", model=model, tokenizer=tokenizer)
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=vectorstore.as_retriever()
)
```

```python
def answer_question(query):
    return qa_chain.run(query)
```

```python
import gradio as gr

# Define the answering function
def answer_question(query):
    return qa_chain.run(query)

# Sample questions
examples = [
    ["What is software engineering?"],
    ["Explain the waterfall model."],
    ["What are the key principles of agile development?"],
    ["Describe the V-model in software engineering."]
]

# Create the interface
with gr.Blocks() as demo:
    gr.Markdown("## 📘 CTSE Lecture Notes Chatbot")
    gr.Markdown("Ask questions.")

    with gr.Row():
        input_text = gr.Textbox(
            lines=4,
            placeholder="Type your question here...",
            label="Your Question"
        )

    submit_button = gr.Button("Get Answer")
    output_text = gr.Textbox(
        lines=5,
        label="Answer",
        interactive=False
    )

    # Link example questions to input
    gr.Examples(examples=examples, inputs=input_text)

    # Connect the button to the function
    submit_button.click(fn=answer_question, inputs=input_text, outputs=output_text)

# Launch the app
demo.launch()
```
